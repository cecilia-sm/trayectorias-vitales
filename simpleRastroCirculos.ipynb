```python
import cv2
import numpy as np

max_value = 255
max_value_H = 360
low_H = 0
low_S = 0
low_V = 0
high_H = max_value_H
high_S = max_value
high_V = 210
```

```python
# Create tracker object

cap = cv2.VideoCapture("matadero1_endb.mp4")

kernel = np.ones((1,1),np.uint8)
kernel2 = np.ones((6,6),np.uint8)
buffer = []
buffer2 = []
nfra = 0
au = 20

max_radius = 5
frame_range = 200
```

```python

while True:
    ret, frame = cap.read()
    f = cap.get(1)
    
    # Extract the dimensions of the original image
    rows, cols, channels = frame.shape
    image = frame[0:rows, 0:cols]
    
    #roi = frame[0:frame.shape[0], 0:frame.shape[1]]
    #roi = frame[200:rows-200, 100:cols-100]
    
   # Apply mask to ignore area outside the background sheet
    mask = np.zeros((image.shape[0], image.shape[1]), np.uint8)
    mask = cv2.rectangle(mask, (90, 150), (cols-90, rows-120), (255,255,255), -1)
    mask_inv = cv2.bitwise_not(mask)
    roi = cv2.bitwise_or(image, image, mask = mask)
    notrack_area = cv2.bitwise_or(image, image, mask = mask_inv)
#   notrack_area = np.stack((notrack_area,)*3, axis=-1)
    
    
    frame_HSV = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)
    frame_threshold = cv2.inRange(frame_HSV, (low_H, low_S, low_V), (high_H, high_S, high_V))
    erosion = cv2.erode(frame_threshold,kernel,iterations = 1)
    dilation = cv2.dilate(erosion,kernel2,iterations = 1)
    contours, _ = cv2.findContours(dilation, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
  
    cropis = []
    
    for cnt in contours: 
        x, y, w, h = cv2.boundingRect(cnt)
        au = int(w/3)
        if w < 400 and h < 400 and w > 2 and h > 2 :
            img_3 = roi[y-au: y+h+au,x-au: x+w+au]
            cropis.append([img_3,[x,y,w,h,f]])
            
    buffer2.append(cropis) 

#     if nfra<2000 :
#         buffer2.append(cropis)  
#     else :
#         buffer2.append(cropis)
#         buffer2.pop(0)
        
    nfra = nfra + 1  
    leni = len(buffer2)

    for x in range(leni):
        lenio = len(buffer2[x])
        for j in buffer2[x]:
            x, y, w, h, fthen = j[1]
            au = int(w/3)
            #roi[y-au:y+h+au, x-au:x+w+au,:3] = j[0]
            #cv2.rectangle(roi, (x, y), (x + w, y + h), (0, 255, 255), 3)
            
            #color shifts from red to blue over time
            diff = f - fthen
            r = 255 - (f - fthen)//(frame_range/255)
            g = 20
            b = 30-(f - fthen)//(frame_range/255)
            color = (b,g,r) 
            
            #radius of circle depends of time since first appeared
            if f == fthen  :
                cv2.rectangle(roi, (x, y), (x + w, y + h), (30, 20, 230), 1)
            else:
                r = 1
#                 decrease = int(max_radius - (f - fthen)//(frame_range/max_radius))
#                 if decrease > 1:
#                     r = decrease
#                 else:
#                     r = 1
                cv2.circle(roi, (x,y), r, color, -1, cv2.LINE_AA) #(80,20,255)
            
    
    # Combine the two images
    output = roi + notrack_area
    output_neg = 255 - roi
    output = cv2.resize(output, (600,900))
    output_neg = cv2.resize(output_neg, (600,900))
    cv2.imshow("Frame", output)
    cv2.imshow("negative", output_neg)
    print(len(buffer2[0]))
    key = cv2.waitKey(30)
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
